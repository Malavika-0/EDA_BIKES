```python
for col in df:
  if df[col].dtype=='int' or df[col].dtype=='float':
    print(col)
    print('---------------')
    print(f'Mean: {df[col].mean()}')
    print(f'Median: {df[col].median()}')
    print(f'Mode: {df[col].mode()}')
    print(f'Minimum of {col}: {df[col].min()}')
    print(f'Maximum of {col}: {df[col].max()}')
    print(f'Variance of {col}: {df[col].var()}')
    print(f'Standard Deviation: {df[col].std()}')
    print()
    plt.figure(figsize=(7,4))
    plt.boxplot(df[col])
    plt.title(col)
    plt.show()
    print('_____________________________________________________________________________')
    print()
```

# **feature engineering**

```python
df.drop(columns=['Sale_ID','Customer_ID','Salesperson_ID'],inplace=True)
```

```python
df
```

```python
df['Customer_Age'].min()
```

```python
df['Customer_Age'].max()
```

```python
df['Age_cat']=df['Customer_Age'].apply(lambda x:'Seniors' if x>=60 else 'Youth' if x<=25 else 'Adults')
```

```python
df
```

```python
# TOTAL SALES
df['Total_Sales'] = df['Price'] * df['Quantity']
print(df[['Bike_Model', 'Total_Sales']].head())
```

```python
df
```

```python
df['Is_weekend'] = df['Date'].dt.dayofweek.isin([5,6]).astype(int)
```

```python
df['Month_Name']=df['Date'].dt.month_name()
```

```python
df['Year']=df['Date'].dt.year
```

```python
df
```

```python
for col in df:
  if df[col].dtype=='object':
    print(df[col].value_counts())
    print("")
    print()
```

```python
#TOTAL COUNT OF EACH MODEL
df['Bike_Model'].value_counts()
```

```python
#TOTAL SALE OF EACH BIKE TYPE bike with highest revenue
df[['Bike_Model','Total_Sales']].groupby('Bike_Model').sum().sort_values(by='Total_Sales',ascending=False)
```

```python
#TOTAL SALES IN EACH LOCATION
df[['Store_Location','Total_Sales']].groupby('Store_Location').sum().sort_values(by='Total_Sales',ascending=False)
```

```python
#AVERAGE PRICE OF EACH MODEL OF BIKE
df[['Bike_Model', 'Price']].groupby('Bike_Model').mean().sort_values(by='Price',ascending=False)
```

```python
#TOTAL SALE BY AGE CATEGORY
df[['Age_cat', 'Total_Sales']].groupby('Age_cat').sum().sort_values(by='Total_Sales',ascending=False)
```

```python
#TOTAL SALES BY FEMALE IN EACH CATEGORY
female_sales=df[df['Customer_Gender']=='Female']
female_sales[['Age_cat', 'Total_Sales']].groupby('Age_cat').sum().sort_values(by='Total_Sales',ascending=False)
```

```python
#TOTAL SALES BY MALE IN EACH CATEGORY
male_sales=df[df['Customer_Gender']=='Male']
male_sales[['Age_cat', 'Total_Sales']].groupby('Age_cat').sum().sort_values(by='Total_Sales',ascending=False)
```

```python
# count of each age category for customers who use cash as a payment method
cash_payments = df[df['Payment_Method'] == 'Cash']
cash_payments['Age_cat'].value_counts()
```

```python
#AVG SALES IN EACH YEAR
df.groupby(df['Date'].dt.year)['Total_Sales'].mean()
```

```python
#which gender uses each payment method the most
df.groupby(['Customer_Gender', 'Payment_Method']).size().unstack(fill_value=0)    #size() function counts the number of entries for each combination of gender and payment method.
                                                                                  # The unstack() method converts the gender into rows and payment methods into columns
```

```python
#which gender uses each payment method the most
df.groupby(['Age_cat', 'Payment_Method']).size().unstack(fill_value=0)
```

```python
#  Weekend vs Weekday sales
df.groupby('Is_weekend')['Total_Sales'].mean()
```

```python
#Group by Bike_Model and calculate average price and quantity sold
df.groupby('Bike_Model').agg({'Price': 'mean', 'Quantity': 'sum'})
```

# visualisation

```python
df.columns
```

```python
#line
df_year=df.sort_values('Year')
```

```python
res=df_year[['Year','Total_Sales']].groupby('Year').agg('mean')
plt.figure(figsize=(10,4))
plt.plot(res.index,res)
plt.title('Year v/s Mean Total Sales')
plt.xlabel('Year')
plt.ylabel('Average Sales')
plt.xticks(range(2020,2025))                #to get all the year yticks() also available
plt.show()
```

```python
#countplot
for col in df:
  if df[col].dtype=='object':
    sns.countplot(data=df,x=df[col])
    plt.grid()
    plt.xticks(rotation=90)
    plt.show()
```

```python
#histplot
plt.figure(figsize=(10,8))
sns.histplot(data=df,x='Customer_Age',bins=6,hue='Bike_Model')
plt.legend()
plt.xticks(rotation=90)
plt.show()
```

```python
#heatmap
df.corr(numeric_only=True)
sns.heatmap(df.corr(numeric_only=True),annot=True)
```

```python
#pie plot
plt.pie(df['Store_Location'].value_counts(),labels=df['Store_Location'].unique(),autopct="%1.0f%%")
plt.title("Store Location value counts")
plt.plot()
```

```python
# Customer age distribution
plt.figure(figsize=(10,6))
sns.histplot(data=df, x='Age_cat')
plt.title('Customer Age Distribution')
```

```python
# 1. Monthly Sales Trend
monthly_sales = df.groupby(['Year', 'Month_Name'])['Total_Sales'].sum().reset_index()

# Plotting the monthly sales trend
plt.figure(figsize=(14, 7))
sns.lineplot(data=monthly_sales, x='Month_Name', y='Total_Sales', hue='Year', marker='o')
plt.title('Monthly Sales Trend (by Year)')
plt.xlabel('Month_Name')
plt.ylabel('Total_Sales ($)')
plt.legend(title='Year')
plt.grid(True)
plt.show()
```

```python
# Revenue by store location
plt.figure(figsize=(10,6))
sns.barplot(data=df, x='Store_Location', y='Total_Sales')
plt.title('Revenue by Store Location')
```

```python
# 2. Total Sales by Bike Model
bike_model_sales = df.groupby('Bike_Model')['Total_Sales'].sum().sort_values(ascending=False).reset_index()

# Plotting total sales by bike model
plt.figure(figsize=(12, 6))
sns.barplot(data=bike_model_sales, x='Total_Sales', y='Bike_Model')
plt.title('Total Sales by Bike Model')
plt.xlabel('Total Sales ($)')
plt.ylabel('Bike Model')
plt.show()
```

```python
# 3. Total Sales by Payment Method
payment_method_sales = df.groupby('Payment_Method')['Total_Sales'].sum().sort_values(ascending=False).reset_index()

# Plotting total sales by payment method
plt.figure(figsize=(10, 6))
sns.barplot(data=payment_method_sales, x='Total_Sales', y='Payment_Method')
plt.title('Total Sales by Payment Method')
plt.xlabel('Total Sales ($)')
plt.ylabel('Payment Method')
plt.grid(True)
plt.show()
```
